```python
sp500 = yf.Ticker("^GSPC") #GSPC - symbol for SNP500 index
```

```python
sp500 = sp500.history(period="max") #Query historical prices
```

```python
del sp500["Dividends"]
del sp500["Stock Splits"]
```

```python
sp500["Tomorrow"] = sp500["Close"].shift(-1) #Creates a column "Tomorrow" on the same day's column based on the next day's closing price
```

```python
sp500
```

```python
sp500["Target"] = (sp500["Tomorrow"] > sp500["Close"]).astype(int) #Setup a target based on tomorrow's price
```

```python
sp500
```

```python
#Remove all data that came before 1990 using Pandas loc method 
#Only take rows where the index is atleast 1990
#.copy() because if you don't, pandas setting with copy warning arises when you try to subset a df and then later assign back to it
sp500 = sp500.loc["1990-01-01":].copy()
```

```python
sp500
```

```python
# Train first ML model
# Using random forest classifier
# It can perform both regression and classification tasks.
# A random forest produces good predictions that can be understood easily.
# It can handle large datasets efficiently.
# The random forest algorithm provides a higher level of accuracy in predicting outcomes over the decision tree algorithm.
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators=100, min_samples_split=100, random_state=1)
#in_sample_split=100 to protect against overfitting, if the tree is built too deep
#n_estimators=100 is the number of individual decision trees we want to train - higher the better accuracy upto a limit (high number makes it slower)
#random_state=1 means if run the same model twice, the random numbers that are generated will be in a predictable sequence each time using this random seed of one 
#meaning if we rerun the model twice, we get the same results, in case you want to update/improve your model to make sure its actually the model

train = sp500.iloc[:-110]
#Last 100 rows not included in the training set to prevent "leakage" from future data and actually train the model to predict the price from past data
test = sp500.iloc[-100:]

predictors = ["Close", "Volume", "Open", "High", "Low"]
model.fit(train[predictors], train["Target"])
```

```python
from sklearn.metrics import precision_score
#what percentage of time did the market go up when we said it's gonna go up?

preds = model.predict(test[predictors])
#generate predictions
```

```python
import pandas as pd
preds = pd.Series(preds, index=test.index)
#convert from numpy array to pandas series - easier to work with that way
```

```python
preds
```

```python
precision_score(test["Target"], preds)
#Calculate the precision score using the actual target and the predicted target
```

```python
#plot our predictions by concatenating the actual values and the predicted values
combined = pd.concat([test["Target"], preds], axis=1)
```

```python
combined.plot()
```

```python
#Results are not satisfying, build another model - do a backtest
#Create a prediction function to enable backtesting, this will wrap up everything we did into one function
def predict(train, test, predictors, model):
  model.fit(train[predictors], train["Target"]) #Fitting of the model using the training predictors and the target
  preds = model.predict(test[predictors]) #Generating our predictions
  preds = pd.Series(preds, index=test.index, name="Predictions")
  combined = pd.concat([test["Target"], preds], axis=1) #Combining our model into a series
  return combined
```

```python
def backtest(data, model, predictors, start = 2500, step = 250): #backtest fn which takes in snp500 data - an ML model, which takes predictors also
#start=2500 when you back test you need to have a certain amount of data to train your first model viz. every trading year has 250 days so 10 years of data, and train your first model using the same
#step=250 we will be training the model for about a year, and then going to next year, and then the next year
#So we'll take the first 10 years of data and predict the value for the 11th year, then first 11 years of data, and predict the 12th year and vice versa
#It gets better by each year, and increases confidence of the model
    all_predictions = [] #create a list called all_predictions where it stores a list of all dataframes where each df will be the prediction for a single year

    for i in range(start, data.shape[0], step): #create a function to loop across our data year by year and make predictions for all the years except the first 10 or so
      #split up our training and testing data
      train = data.iloc[0:i].copy() #all of the years prior to the current year
      test = data.iloc[i:(i+step)].copy()# test is the current year.copy() to avoid copy warning
      predictions = predict(train, test, predictors, model) #use predict function to generate our predictions
      all_predictions.append(predictions) #append to all predictions, append our predictions to the given year
    return pd.concat(all_predictions) #concatenate all our predictions into a single df
```

```python
predictions = backtest(sp500, model, predictors) #backtest sp500 data with the model and predictors we created earlier
```

```python
#Evaluate error in predictions
predictions["Predictions"].value_counts() #taking a look at predictions and see how many days we predicted the market would go up vs down
#value_counts() will just count up how many times each type of predictions made
```

```python
#0 is how many days the market would go down which is 3431 days
#1 is how many days the market would go up which is 2492 days
precision_score(predictions["Target"], predictions["Predictions"]) #checking precision score
```

```python
predictions["Target"].value_counts()/predictions.shape[0] #percentage of days when the market actually went up - so we look the value_counts() of the target divided by the no. rows
```

```python
#Add more predictors into the model and see if that improves the accuracy
#Say if you are a human analyst trying to figure out if the stock price would go up tomorrow
#some of the numbers you might look at are is the stock price today higher than it was last week
#than it was three months ago, a year ago

#horizons are horizons on which we want to look at the rolling means so we'll calculate
#the mean close price in the last two days, the last trading week (5), the last 60 mmonths
#the last year, and then the last 4 years and we'l find the ration between today's 
#closing price and the closing price in those periods which will help us know 
#whether the market looks bullish, and if so it may be due for a downturn 
```

```python
horizons = [2,5,60,250,1000] #Creates variety of rolling averages
new_predictors = [] #will hold new columns that are created

for horizon in horizons: #loop through these horizons
    rolling_averages = sp500.rolling(horizon).mean() #calculate RA against that horizon, and take the mean
    
    ratio_column = f"Close_Ratio_{horizon}" #Create Ration columns
    sp500[ratio_column] = sp500["Close"] / rolling_averages["Close"] #add that column to the sp500 dataframe
    #first loop will the ration b/w today's close and the avg close in the last 2 days
    #second loop will be the ratio b/w today's close and te avg close in the last 5 days
    #and vice versa
    trend_column = f"Trend_{horizon}" #no. of days in the past x days whatever the horizon is that the stocl price actually went up
    sp500[trend_column] = sp500.shift(1).rolling(horizon).sum()["Target"] #shift forward and find the rolling sum of the target
    
    new_predictors+= [ratio_column, trend_column]
```

```python
sp500 = sp500.dropna(subset=sp500.columns[sp500.columns != "Tomorrow"]) #get rid of the extra rows that has NaN values
```

```python
model = RandomForestClassifier(n_estimators=200, min_samples_split=50, random_state=1)
```

```python
def predict(train, test, predictors, model):
    model.fit(train[predictors], train["Target"])
    preds = model.predict_proba(test[predictors])[:,1] #predict.proba is the probability that the row will be a zero or a one
    #basically to see if the 
    preds[preds >=.6] = 1 #default threshold is 50%, set custom threshold to 60%
    #only when price goes up
    preds[preds <.6] = 0 #reduces the no. of trading days but increases the chance that the price will go up on those days
    preds = pd.Series(preds, index=test.index, name="Predictions")
    combined = pd.concat([test["Target"], preds], axis=1)
    return combined
```

```python
predictions = backtest(sp500, model, new_predictors)
```
